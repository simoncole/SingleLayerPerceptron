Import modules

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Dense
```

Load and split data, then check the number of data

```python
data = np.loadtxt('/content/drive/MyDrive/and.csv', delimiter=',', skiprows=1)
data_train = data[:,:-1]
class_train = data[:, -1]
print('data:', data_train.shape)
print('class:', class_train.shape)
```

Model building and learning

```python
model=Sequential()
model.add(Dense(1, input_shape=(2,), activation='sigmoid'))
model.compile(optimizer='sgd', loss='binary_crossentropy', metrics=['accuracy'])
fit_log = model.fit(data_train, class_train, epochs=1000, batch_size=1)
```

Graph for accuracy

```python
plt.plot(fit_log.history['accuracy'], label='train')
plt.title = 'Model accuracy'
plt.xlabel = 'epochs'
plt.ylabel = 'accuracy'
plt.legend()
plt.show()
```

Find the predicted value for the training data

```python
pred = model.predict(data_train)
print(pred)
```
